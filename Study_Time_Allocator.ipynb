# **Study Time Allocator**

This program distributes daily study hours across subjects based on their difficulty level.
Harder subjects get more time, while easier ones get less time.
The goal is to help organize study time in a balanced and efficient way.

Project developed by: Reuof Bahmid


```python
subjects = int(input("How many hours do you have?"))
study_hours = float(input("How many hours can you study per day?" ))
names = []
difficulties =[]

for i in range(subjects):
  name = input(f"Enter subhject {i+1} name: ")
  difficulty =int(input("Difficulty (1_5):"))
  names.append(name)
  difficulties.append(difficulty)

total_difficulty = sum(difficulties)

print("📚🕦 Your study plan:")

for i in range(subjects):
  hours = (difficulties[i] / total_difficulty) * study_hours

  h = int(hours)
  m = int((hours - h)*60)
  print(f"{names[i]}  \u200e -> {h} hours and {m} minutes per day ")
```
